```python
file1 = pd.read_csv('tweets-data/anies.csv', delimiter=";")
file2 = pd.read_csv('tweets-data/ganjar.csv', delimiter=";")
file3 = pd.read_csv('tweets-data/prabowo.csv', delimiter=";")

combined_df = pd.concat([file1, file2, file3], ignore_index=True)
combined_df.to_csv('combined_file.csv', index=False)

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('combined_file.csv')

# Display the DataFrame
display(df)
```

```python
full_text_column = df['full_text']
full_text_df = df[['full_text']]
```
