```python
file1 = pd.read_csv('tweets-data/aniesafterdebat.csv', delimiter=",")
file2 = pd.read_csv('tweets-data/ganjarafterdebat.csv', delimiter=",")
file3 = pd.read_csv('tweets-data/prabowoafterdebat.csv', delimiter=",")

combined_df = pd.concat([file1, file2, file3], ignore_index=True)
combined_df.to_csv('combined_file.csv', index=False)

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('combined_file.csv')

# Display the DataFrame
display(df)
```
